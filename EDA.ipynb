####  value_counts() function only works on pandas series

```python
df['drive-wheels'].value_counts()
```

```python
#converting the series to data frame.
df['drive-wheels'].value_counts().to_frame()
```

```python
drive_wheels_count = df['drive-wheels'].value_counts().to_frame()
drive_wheels_count.rename(columns = {"drive-wheels": 'value_counts'}, inplace=True)
drive_wheels_count
```

```python
drive_wheels_count.index.name = 'drive-wheels'
drive_wheels_count
```

```python
engine_location_counts = df['engine-location'].value_counts().to_frame()
engine_location_counts.rename(columns = {"engine-locatioin":"values_count"}, inplace=True)
engine_location_counts.index.name = 'engine-location'
engine_location_counts
```

<p>Examining the value counts of the engine location would not be a good predictor variable for the price. This is because we only have three cars with a rear engine and 198 with an engine in the front, this result is skewed. Thus, we are not able to draw any conclusions about the engine location.</p>

####  ANOVA Test

steps to perform anova test
1. prepare the data using {group by} 
2. passed grouped data to f_oneway from stats moduel of scipy

```python
grouped_test2 = df[['drive-wheels', 'price']].groupby(['drive-wheels'])
grouped_test2.head(2)
```

```python
grouped_test2.get_group('4wd')['price']
```

```python
df['drive-wheels'].value_counts()
```

```python
fval, pval = s.f_oneway(grouped_test2.get_group('fwd')['price'], grouped_test2.get_group('rwd')['price'], grouped_test2.get_group('4wd')['price'])
print('ANOVA result is F= ', fval, ' P= ', pval)
```
